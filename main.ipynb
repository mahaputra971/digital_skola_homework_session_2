### 1. Gathering Data

```python
'''
Membaca data dari file CSV
'''
df_1 = pd.read_csv('dataset_hw_2/lampiran_1.csv')
```

### 3. Data Transform

```python
''' 
mengubah nama tabel dengan menambahkan satuan agar menghindari bias data
'''
df_1.rename(columns={
    'gaji' : 'gaji_idr',
    'berat': 'berat_kg',
    'tinggi': 'tinggi_cm',
}, inplace=True)
```

```python
''' 
Menghapus data yang duplikat pada Dataframe
'''
df_1_no_duplicated = df_1.drop_duplicates()
df_1_no_duplicated.duplicated().sum()
```

```python
''' 
Mengecheck kembali data yang hilang sebelum dihapus
'''
df_1_no_duplicated.isnull().sum()
```

```python
''' 
Menghapus baris di kolom yang memiliki nilai null/nan yang mana sulit didapatkan alternatif untuk mengisi data null/nan tersebut
''' 
df_1_no_dup_no_null = df_1_no_duplicated.dropna(subset=['umur', 'jenis_kelamin', 'gaji_idr', 'berat_kg', 'tinggi_cm', 'is_merokok', 'is_menikah', 'pendidikan', 'ever_stress'])
df_1_no_dup_no_null.isnull().sum()
```

```python
'''
Mengisi data kosong yang ada pada kolom 'pekerjaan' dengan string 'Pengangguran' karena menganggap data yang kosong dianggap tidak memiliki pekerjaan
'''
df_1_no_dup_no_null.loc[:, 'pekerjaan'] = df_1_no_dup_no_null['pekerjaan'].fillna('Pengangguran')
df_1_no_dup_no_null.isnull().sum()
```

```python
''' 
Mengecheck kembali info umum dari Dataframe yang telah dibersihkan
'''
df_1_no_dup_no_null.info()
```

```python
''' 
Mengubah datatype dari kolom 'is_menikah' dan 'is_merokok' dari FLOAT ke INT karena hanya memiliki nilai 0 atau 1
'''
df_1_no_dup_no_null['is_menikah'] = df_1_no_dup_no_null['is_menikah'].astype(int)
df_1_no_dup_no_null['is_merokok'] = df_1_no_dup_no_null['is_merokok'].astype(int)

print(df_1_no_dup_no_null.dtypes)
```

```python
''' 
Mengecheck statistik deskriptif dari Dataframe
'''
df_1_no_dup_no_null.describe()
```

```python
''' 
Insight yang didapat : 
1. Data minimal dari tabel 'umur' adalah '-1' yang mana itu tidak masuk akal
2. Data lainnya memiliki nilai yang masuk akal
'''
```

```python
''' 
Menentukan outlier pada kolom 'umur' dengan menggunakan metode IQR (Interquartile Range)
'''
# Q1, Q3, dan IQR  
Q1 = df_1_no_dup_no_null['umur'].quantile(0.25)
Q3 = df_1_no_dup_no_null['umur'].quantile(0.75)
IQR = Q3 - Q1

# Batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter baris yang berada di luar batas
outliers_lower = df_1_no_dup_no_null[(df_1_no_dup_no_null['umur'] < lower_bound)]
outliers_higher = df_1_no_dup_no_null[(df_1_no_dup_no_null['umur'] > upper_bound)]
```

```python
''' 
Melihat data outlier lower bound pada kolom 'umur'
'''
outliers_lower = outliers_lower.sort_values(by='umur', ascending=True)
outliers_lower
```

```python
'''
Insight yang didapat dari data outlier lower bound pada kolom 'umur':
1. Terdapat 6 baris data yang memiliki umur -1 yang mana itu tidak masuk akal
'''
```

```python
''' 
Melihat data outlier higher bound pada kolom 'umur'
'''
outliers_higher = outliers_higher.sort_values(by='umur', ascending=False)
outliers_higher
```

```python
''' 
Menghilangkan data umur yang tidak masuk akal yaitu umur -1
'''
df_1_no_dup_no_null = df_1_no_dup_no_null[df_1_no_dup_no_null['umur'] >= 0]
df_1_no_dup_no_null.describe()
```

```python
''' 
Mengecheck kembali info umum dari Dataframe yang telah dibersihkan
'''
df_1_no_dup_no_null.info()
```

### 4. Export Clean Data

```python
''' 
Melakukan ekspor Dataframe yang telah dibersihkan ke file CSV dengan catatan : 
1. Datatype kolom 'umur' tetap FLOAT karena ada kemungkinan umur yang ditentukan memang tidak bulat (contoh : 22.5 tahun)
2. Ada beberapa data anomali seperti umur 2 tahun sudah pendidikan S1, dan sebagainya
'''
df_1_no_dup_no_null.to_csv('clean_data/clean_data.csv', index=False)
```
